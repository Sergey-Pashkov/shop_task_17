### Домашнее задание: Создание шаблонов для интернет-магазина

#### Задача:

Вы будете разрабатывать шаблоны для простого интернет-магазина, который будет отображать список товаров и детали каждого товара. Используйте знания о шаблонах, фильтрах и наследовании.

#### Этапы выполнения:

1. **Создание модели товара**:
   - Создайте модель `Product` в `models.py`, которая будет содержать следующие поля:
     - `name` (название товара, CharField)
     - `description` (описание товара, TextField)
     - `price` (цена товара, DecimalField)
     - `image` (изображение товара, ImageField)
     - `created_at` (дата создания, DateTimeField)

2. **Создание представлений**:
   - Создайте два представления:
     - `product_list`: отображает список всех товаров. Передайте контекст с информацией о товарах в шаблон.
     - `product_detail`: отображает детальную информацию о конкретном товаре. Используйте `product_id`, чтобы получить нужный товар из базы данных.

3. **Создание шаблонов**:
   - Создайте базовый шаблон `base.html`, который будет содержать общий стиль (например, хэдер и футер).
   - Создайте шаблон `product_list.html`, который будет наследоваться от `base.html` и отображать список товаров. Каждый товар должен содержать название, цену и изображение. Используйте фильтры для форматирования цены (например, отображать ее с двумя знаками после запятой).
   - Создайте шаблон `product_detail.html`, который также будет наследоваться от `base.html` и отображать детальную информацию о товаре, включая название, описание, цену и изображение.

4. **Статические файлы**:
   - Используйте статические файлы для стилизации ваших шаблонов (например, добавьте CSS для улучшения внешнего вида).

5. **Рендеринг данных**:
   - Убедитесь, что все данные корректно рендерятся в шаблонах. Используйте конструкции `{% for %}`, `{% if %}` и фильтры.

#### Дополнительные требования:

- **Документация**: Добавьте комментарии в код, объясняющие основные моменты реализации.
- **Обработка ошибок**: Добавьте обработку случая, когда не найден товар, на странице `product_detail.html`.
- **Структура проекта**: Поддерживайте организованную структуру каталогов для шаблонов, используя стандартные практики.

- **Качество и читаемость кода** (30%)
- **Эстетика и функциональность интерфейса** (20%)
